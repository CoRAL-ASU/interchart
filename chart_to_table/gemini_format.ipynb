```python
import os
import json
```

```python
types = ['t1', 't2', 't3']
```

```python
for type in types:
	all_charts = json.load(open(f'./{type}_charts.json'))
	
	for key, value in all_charts.items():
		tables = []

		for chart in value:
			chart_name = chart.split('.')[0]
			data = json.load(open(f'../../model_responses/chart_to_table/gemini/{type}/{chart_name}.json'))
			tables.append(data[0])

		os.makedirs(f'./{type}_tables', exist_ok=True)
		with open(f'./{type}_tables/{key}.json', 'w') as f:
      			json.dump(tables, f, indent=4)
```
