```python
import os
import json
import pandas as pd
```

```python
type_1_df = pd.read_json('../../type1/qa.json')
type_2_df = pd.read_json('../../type2/qa_0.json')
type_3_df = pd.read_json('../../type3/qa.json')
```

```python
t1_chart_names = [f'multichart_{i}_' for i in type_1_df['image_index'].unique()]
t2_chart_names = [f'{i}_' for i in type_2_df['chart_name'].unique()]
t3_chart_names = [f'{i}_' for i in type_3_df['image_index'].unique()]
```

```python
t1_charts = {}
t2_charts = {}
t3_charts = {}

all_t1_charts = os.listdir('../../type1/simple')
all_t2_charts = os.listdir('../../type2/simple')
all_t3_charts = os.listdir('../../type3/simple')

for chart_name in t1_chart_names:
    t1_charts[chart_name[:-1]] = sorted([chart for chart in all_t1_charts if chart.startswith(chart_name)])
    
for chart_name in t2_chart_names:
    t2_charts[chart_name[:-1]] = sorted([chart for chart in all_t2_charts if chart.startswith(chart_name)])
    
for chart_name in t3_chart_names:
    t3_charts[chart_name[:-1]] = sorted([chart for chart in all_t3_charts if chart.startswith(chart_name)])
    
```

```python
json.dump(t1_charts, open('t1_charts.json', 'w'), indent=4)
json.dump(t2_charts, open('t2_charts.json', 'w'), indent=4)
json.dump(t3_charts, open('t3_charts.json', 'w'), indent=4)
```

```python
type2_old = pd.read_json('../../type2/qa_1.json')
type2_new = pd.read_json('../../type2/qa_0.json')

type2_old_chart_names = set(type2_old['chart_name'].unique())
type2_new_chart_names = set(type2_new['chart_name'].unique())

left_charts = type2_old_chart_names - type2_new_chart_names
```

```python
all_t2_charts = os.listdir('../../type2/simple')

left_charts_dict = {}
for chart_name in left_charts:
	left_charts_dict[chart_name] = sorted([chart for chart in all_t2_charts if chart.startswith(chart_name)])
 
json.dump(left_charts_dict, open('t2_left_charts.json', 'w'), indent=4)
```
