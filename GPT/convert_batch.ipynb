```python
import pandas as pd
```

```python
df = pd.read_csv('list_of_batches.csv')
```

```python
names = df['metadata'].values
ids = df['id'].values
```

```python
names[-1], ids[0]
```

```python
batch_ids = {}

datasets, imgs, prompts = set(), set(), set()

for name, id in zip(names, ids):
    dataset, img, prompt, _, idx = name.split('_')
    
    datasets.add(dataset)
    imgs.add(img)
    prompts.add(prompt)
    
    if prompt not in batch_ids:
        batch_ids[prompt] = {}
    if dataset not in batch_ids[prompt]:
        batch_ids[prompt][dataset] = {}
    if img not in batch_ids[prompt][dataset]:
        batch_ids[prompt][dataset][img] = [None]*70
    batch_ids[prompt][dataset][img][int(idx)] = id
    
```

```python
for prompt in batch_ids:
    for dataset in batch_ids[prompt]:
        for img in batch_ids[prompt][dataset]:
            # remove None from end
            batch_ids[prompt][dataset][img] = [id for id in batch_ids[prompt][dataset][img] if id is not None]
```
