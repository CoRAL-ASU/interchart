```python
from openai import OpenAI
import os
import json
from concurrent.futures import ThreadPoolExecutor 
client = OpenAI(api_key = "<api_key>")
```

```python
batch_ids = json.load(open("batch_ids.json"))
```

```python
def get_status(batch_id):
	return client.batches.retrieve(batch_id).status
```

```python
completed_count = 0
not_completed = 0

for prompt_type in ['zeroshot', 'zeroshotcot', 'directives']:
	for dataset in ['type1', 'type3']:
		for image_type in ['simple', 'original', 'combined']:
			if image_type == 'original' and dataset == 'type3':
				continue

			with ThreadPoolExecutor(max_workers=50) as executor:
				statuses = list(executor.map(get_status, batch_ids[prompt_type][dataset][image_type]))

			for status in statuses:
				if status == "completed":
					completed_count += 1
				else:
					not_completed += 1
```
